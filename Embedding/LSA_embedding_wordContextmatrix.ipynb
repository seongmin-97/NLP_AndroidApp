```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from soynlp.vectorizer import sent_to_word_contexts_matrix
```

```python
DATA_IN_PATH = '../data/preprocess/'
# 데이터 불러오기 !
naverReview = pd.read_csv(DATA_IN_PATH + 'naverReview_preprocess_Okt.csv', encoding='UTF8')
```

```python
print(naverReview.shape) # 출력 결과 20만 개의 데이터
naverReview.head()
```

```python
#데이터를 하나의 데이터 프레임으로 합치자!
data = naverReview
```

```python
corpus = [str(context) for context in data['context']] # 데이터들을 리스트 형태로 저장
```

```python
# 단어-문맥 행렬 생성, idx2vocab은 단어 들이 모인 리스트 인듯
input_matrix, idx2vocab = sent_to_word_contexts_matrix(corpus, windows=3, min_tf=10, dynamic_weight=True, verbose=True)
```

```python
type(input_matrix) # 타입이 뭘까 궁금해서 한번...
```

```python
input_matrix # 26870개의 단어이므로 26870*26870 의 행렬이다.
```

```python
type(idx2vocab)
```

```python
idx2vocab
```

```python
cooc_svd = TruncatedSVD(n_components=300) # 300차원으로 줄이자!
cooc_vecs = cooc_svd.fit_transform(input_matrix) # 아까 받은 input_matrix를 이 모델에 적용
```

```python
np_cooc_vecs = np.array(cooc_vecs) # 넘파이로 저장하자
np.save(open('./embedding_data/LSA_embedding_wordContextmatrix.npy', 'wb'), np_cooc_vecs)
```
