```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
```

```python
DATA_IN_PATH = '../data/preprocess/'

naverReview = pd.read_csv(DATA_IN_PATH + 'naverReview_preprocess_Okt.csv', encoding = 'UTF8')
```

```python
data = naverReview
#데이터 저장
```

```python
contexts = list(data['context'])

sentences = []
for context in contexts :
    sentences.append(str(context).split())
#데이터 변환
```

```python
sentences
```

```python
model = Word2Vec(sentences, size=100, workers=4, sg=1)
model.save('./embedding_data/Word2Vec_embedding.model')
#Word2Vec으로 임베딩
```
