```python
import numpy as np
import pandas as pd
import re
from konlpy.tag import Okt
```

```python
DATA_IN_PATH = './'
```

```python
train_data = pd.read_csv(DATA_IN_PATH + 'ratings_train (1).txt', header=0, delimiter='\t', quoting=3 )
train_data
#(1)좀 거슬리네 이게 정말 잘 정리된 파일이라는걸 새삼 느꼈어
```

```python
def preprocessing(review, okt, remove_stopwords = True):

    review_text = re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]", "", review)
    word_review = okt.morphs(review_text, stem=True)
    stop_words = set(['\r\n ', '\r\n\t\t ', '\r\n\r\n ', '\r\n\t\t', '\r\n', '\r\n\t\t\r\r\n\t\t', '\r\n\r\n\r\n\r\n\r\n\t\t', '\r\n\r\n', '\r\n\r\n\r\n', '\r\n\r\n\t\t', '\r\n\r\n\r\n\t\t', '\r\n\r\n\r\n\r\n\t\t', '\r', '\n', '\t', '은', '는', '이', '가', '하', '아', '것', '들','의', '있', '되', '수', '보', '주', '등', '한'])
    
    if remove_stopwords:
        word_review = [token for token in word_review if not token in stop_words]
    return word_review
```

```python
okt = Okt()
clean_train_review = []

for review in train_data['document']:
    if type(review) == str:
        clean_train_review.append(preprocessing(review, okt, remove_stopwords = True))
    else:
        clean_train_review.append([])
```

```python
clean_train_review
```

```python
test_data = pd.read_csv(DATA_IN_PATH + 'ratings_test (1).txt', header=0, delimiter='\t', quoting=3 )
```

```python
test_data
```

```python
okt = Okt()
clean_test_review = []

for review in test_data['document']:
    if type(review) == str:
        clean_test_review.append(preprocessing(review, okt, remove_stopwords = True))
    else:
        clean_test_review.append([])
```

```python
clean_test_review
```

```python
np.save(open('Pretreatmented_ratings_train.npy', 'wb'), clean_train_review)
np.save(open('Pretreatmented_ratings_test.npy', 'wb'), clean_test_review)
#저장
```
